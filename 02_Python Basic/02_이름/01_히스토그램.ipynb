```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## 히스토그램(Histrogram) 개요

* 히스토그램(histrogram)은 데이터의 분포를 시각화하여 나타내는 하나의 방법임.
* 가로축에는 전체 데이터 값을 급간으로 나누어 표시하고 세로축에는 해당 데이터 값의 빈도를 표시함.
* 데이터의 양에 비해 너무 많거나 적은 급간을 사용하면 의미 있는 정보를 얻기 어려움.
* 일반적으로 5개에서 15개 정도의 급간으로 나누게 되며 데이터가 클수록 더 많은 급간을 사용함.

## 방법1: Matplotlib 라이브러리

```python
df = pd.read_csv("dataset/Iris.csv")
df.head()   
```

```python
fig, ax = plt.subplots(figsize = (9, 5))

ax.hist(df["SepalLengthCm"]);
```

* 히스토그램은 색상을 변경하거나 제목 등을 추가하기 위해 다양한 인자를 사용할 수 있음.

```python
fig, ax = plt.subplots(figsize = (12, 7))

ax.hist(df["SepalLengthCm"],
        bins = 20,               # 급간(bin) 개수 설정
        density=True,            # y축 값을 밀도(density)로 표시
        facecolor = "green",     # 색상 설정
        alpha = 0.5)             # 색상 투명도 설정


ax.set(title = "Histogram of Sepal Length",    # 그래프 제목 설정
       xlabel = "Bins",                        # X축 설정
       ylabel = "Density");                    # y축 설정
```

## 방법2: Snsborn 라이브러리

```python
fig, ax = plt.subplots(figsize = (12, 7))

sns.distplot(df["SepalLengthCm"],               
             bins = 20,                        # 급간(bins) 설정
             kde = True,                       #  Kernel Density Estimate 그래프 
             color = "green");                 # 색상 설정
            
ax.set(title = "Histogram of Sepal Length",    # 그래프 제목 설정
       xlabel = "Bins",                        # X축 설정
       ylabel = "Density");                    # y축 설정
```
