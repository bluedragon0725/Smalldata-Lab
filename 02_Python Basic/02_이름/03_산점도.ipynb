```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = sns.load_dataset("iris")
df.head()
```

## 산점도(Scatter plot) 개요

* 산점도(scatter plot)는 동일한 개체에 대해 측정한 두 개의 양적 변수 사이의 관계를 나타내는데 가장 유용한 그래프임.
* 한 변수의 값은 수평축에, 다른 변수의 값은 수직축에 나타내며, 일반적으로 설명변수를 수평축(x축)으로 반응변수를 수직축(y축)으로 설정함.
* 두 변수 간의 전반적인 관계(overall relationship)을 살펴보고 해당 패턴에서 눈에 띄게 벗어난 이탈 현상에 주목함.
* 두 변수간 관계의 방향, 형태, 강도를 바탕으로 산점도의 전반적인 패턴을 설명하게 되며, 전반적인 패턴 밖에 위치하는 이상치(outlier)에 주목함.


## 방법1: Matplotlib

* 별도로 x, y를 지정하는 않는 경우, 괄호 안에서 처음 나오는 객체가 x 좌표, 그 다음 나오는 객체가 y 좌표로 인식됨.

```python
fig, ax = plt.subplots()

ax.plot(df["sepal_length"], df["sepal_width"],       # x축, y축 지정
            marker = "o",                            # 점(point) 기호 유형 
            linestyle = "none",                      # 선 그래프 유형
            color = "red",                           # 색상
            alpha = 0.5);                            # 색상 투명도

ax.set_title("Sepal_length vs Sepal_width")
ax.set_xlabel("Sepal_length")
ax.set_ylabel("Sepal_width");
```

* 산점도에 사용되는 점(point) 기호 유형을 아래 그림과 같이 다양하게 설정할 수 있음. [(출처:matplotlib.org)](https://matplotlib.org/3.1.3/gallery/lines_bars_and_markers/marker_reference.html#sphx-glr-gallery-lines-bars-and-markers-marker-reference-py)

 * linestyle 인자를 사용하여 산점도에 아래와 같이 다양한 유형의 선(line)을 추가할 수 있음. [(출처:matplotlib.org)](https://matplotlib.org/3.1.3/gallery/lines_bars_and_markers/linestyles.html#sphx-glr-gallery-lines-bars-and-markers-linestyles-pyy)


<img src="https://matplotlib.org/3.1.3/_images/sphx_glr_linestyles_001.png">

* axhline( ) 함수에서 linestyle 인자로 선 유형을 지정할 수 있으며 선의 두께는 linewidth 인자를 이용하여 설정이 가능함.

```python
fig, ax = plt.subplots()

ax.plot(df["sepal_length"], df["sepal_width"],       # x축, y축 지정
            marker = "o",                            # 점(point) 기호 유형 
            linestyle = "none",                      # 선 그래프 유형
            color = "red",                           # 색상
            alpha = 0.5);                            # 색상 투명도

ax.axhline(y = 3.5, linestyle = "dotted", linewidth = 2, color = "blue")   

ax.set_title("Sepal_length vs Sepal_width", fontsize = 15)
ax.set_xlabel("Sepal_length")
ax.set_ylabel("Sepal_width");
```

## 방법2: Seaborn

```python
fig, ax = plt.subplots()

sns.scatterplot(x = "sepal_length", 
                y = "sepal_width",
                data = df,
                color = "red",
                alpha = 0.5);

ax.axhline(y = 3.5, linestyle = "dotted", linewidth = 2, color = "blue")   

ax.set_title("Sepal_length vs Sepal_width", fontsize = 15)
ax.set_xlabel("Sepal_length")
ax.set_ylabel("Sepal_width");
```
